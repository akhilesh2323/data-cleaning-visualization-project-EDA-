```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
data = pd.read_csv("/kaggle/input/datasets/akhileshyadav331/student-score/score_updated.csv")

# First 5 Rows
print("First 5 Rows:")
display(data.head())

# Dataset Information
print("\nDataset Info:")
data.info()

# Shape
print("\nDataset Shape:")
print(data.shape)

# Missing Values
print("\nMissing Values:")
print(data.isnull().sum())

# Duplicate Values
print("\nDuplicate Values:")
print(data.duplicated().sum())

# Remove Duplicates (if any)
data = data.drop_duplicates()

# Statistical Summary
print("\nStatistical Summary:")
display(data.describe())
```

```python
plt.figure(figsize=(8,5))

plt.hist(data['Scores'],
         bins=20,
         edgecolor='black')

plt.title("Distribution of Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(data['Hours'],
            data['Scores'])

plt.title("Hours Studied vs Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Scores")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.heatmap(
    data.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
sns.pairplot(data)

plt.show()
```

```python
print("EDA INSIGHTS")

print("- Dataset contains", len(data), "records")

print("- Average Score:",
      round(data['Scores'].mean(),2))

print("- Maximum Score:",
      data['Scores'].max())

print("- Minimum Score:",
      data['Scores'].min())

correlation = data['Hours'].corr(data['Scores'])

print("- Correlation between Hours and Scores:",
      round(correlation,2))
```
